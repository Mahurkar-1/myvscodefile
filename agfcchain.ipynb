```python
#load data 
loader = TextLoader("solar_system.txt")
docs = loader.load()
```

```python
#split text into chunks 
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
chunks = text_splitter.split_documents(docs)
```

```python
#Embedding query
embedding = GoogleGenerativeAIEmbeddings(model="models/embedding-001",google_api_key=os.environ["GOOGLE_API_KEY"])
```

```python
#store in vectore db 
db = Chroma.from_documents(chunks, embedding)
```

```python
# Initialize gemini chat model 
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash",temperature=0)
```

```python
# Build qa chain 
qa = RetrievalQA.from_chain_type(llm=llm, retriever=db.as_retriever(), chain_type="stuff")
```
