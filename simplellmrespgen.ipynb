```python
##üëâ ‚ÄúUse any LLM, ask a question and show me the generated response.‚Äù
import os 
from langchain_google_genai import ChatGoogleGenerativeAI

#llm load karna
llm=ChatGoogleGenerativeAI(model="gemini-1.5-flash",google_api_key=os.environ["GOOGLE_API_KEY"])

#set your question 
question="what is langchain why it is important"

#LLM se responce lo
responce=llm.invoke(question)

#output print karo 
print("Question:",question)
print("Answer:",responce.content)
```
