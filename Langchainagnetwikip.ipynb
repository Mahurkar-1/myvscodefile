```python
# step:-1 LangChain & dependencies 
# pip install langchain langchain_community langchain_google_genai langchain_ollama chromadb

# step:-2 Import Required classes
import os
from langchain.agents import initialize_agent, AgentType, Tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.utilities import WikipediaAPIWrapper

# step:-3 Define Tools 
search = WikipediaAPIWrapper()

tools = [
    Tool(
        name="wikipedia",
        func=search.run,
        description="Useful for looking up general knowledge from Wikipedia"
    ),
    Tool(
        name="Calculator",
        func=lambda q: str(eval(q)),
        description="Useful for doing math calculations"
    )
]

# step:-4 Load LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    temperature=0
)

# step:-5 Initialize Agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# step:-6 Run the Agent 
query = "calculate 25*2"
response = agent.invoke({"input": query})
print(response)



#print("ðŸ¤– Ask me something!")
#while True:
    #query = input("You: ")
    #if query.lower() in ["exit", "quit", "bye"]:
        #print("Bot: Goodbye ðŸ‘‹")
        #break
    #response = agent.invoke(query)
    #print("Bot:", response
```
