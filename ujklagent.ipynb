```python
#step:-1 Langchain & dependencies pip install langchian, langchain_community, langchain_googel_genai, langchain_openai, chroma

#step:-2 Import Required classes 
import os 
from langchain.agents import initialize_agent, AgentType, Tool
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.llms import OpenAI
from langchain.tools import DuckDuckGoSearchRun 

#step:-3 Define Tools 
search = DuckDuckGoSearchRun()

tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Useful for searching the internet to find information."
    )
]

#step:-4 Load LLM
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash",temperature=0)

#step:-5 Initialize Agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

#step:-6 Run the Agent
query = "who is the Prime Minister of japan?"
responce = agent.invoke(query)
print(responce)
```
